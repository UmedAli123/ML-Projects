# Data preprocessing

```python
x=wine.drop('target',axis=1)
y=wine['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=10/33,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st_scale=StandardScaler()
train_scale=st_scale.fit_transform(x_train)
test_scale=st_scale.transform(x_test)
```

# MODEL 1 USING SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svm_model=SVC()
svm_model.fit(x_train,y_train)
```

```python
y_pred=svm_model.predict(x_test)
print("Accuracy of the model is: ",accuracy_score(y_test,y_pred).round(2)*100)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
