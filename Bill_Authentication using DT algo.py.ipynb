```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
bill_auth=pd.read_csv('bill_authentication.csv')
bill_auth.head(8)
```

```python
bill_auth.shape
```

```python
bill_auth.isnull().sum()
```

```python
bill_auth.Class.value_counts()
```

```python
x=bill_auth.drop('Class',axis=1)
y=bill_auth.Class
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=3/10,random_state=42)
print(x.shape, x_train.shape, x_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

```python
DT_Model=DecisionTreeClassifier()
```

```python
DT_Model.fit(x_train,y_train)
```

```python
y_pred=DT_Model.predict(x_test)
print('Accuaracy of the model on testing data is: ',accuracy_score(y_test,y_pred))
```

```python
y_pred=DT_Model.predict(x_train)
print('Accuaracy of the model on training data is: ',accuracy_score(y_train,y_pred))
```

```python
plt.figure(figsize=(30,15))
model_tree=tree.plot_tree(DT_Model,filled=True)
model_tree
plt.show()
```
